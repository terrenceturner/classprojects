# Lab 5 Word Cloud ⭐️

The purpose of this project is to generate a word cloud image file based on RomeoAndJuliet.txt with default background_color and mask_image being mask_star.png

### Visualizing Word Frequencies with Word Clouds

```python
from pathlib import Path

text = Path('RomeoAndJuliet.txt').read_text()
```

```python
import imageio

mask_image = imageio.imread('mask_star.png')
```

```python
from wordcloud import WordCloud

wordcloud = WordCloud(colormap='prism', mask=mask_image,
                     background_color='white')
```

```python
wordcloud = wordcloud.generate(text)
```

```python
wordcloud = wordcloud.to_file('RomeoAndJulietStar.png')
```
